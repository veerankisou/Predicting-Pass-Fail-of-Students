```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
np.random.seed(0)

hours_studied = np.random.uniform(1,10,100)
marks_obtained = 2 * hours_studied + np.random.normal(0,1,100) * 2 + 20

pass_fail = np.where(marks_obtained > 40, 10, 0)

data = pd.DataFrame({
    "Hours Studied" : hours_studied,
    "Marks Obtained" : marks_obtained,
    "Pass/Fail" : pass_fail
})

x = data[["Hours Studied"]]
y = data["Pass/Fail"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_pred_linear = linear_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred_linear)

r2 = r2_score(y_test, y_pred_linear)

print(f"Linear Regression - Mean Squared Error: {mse:.2f}")
print(f"Linear Regression - R^2 score: {r2:.2f}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_test, y_test, color = "blue", label = "Actual Marks")
plt.plot(x_test, y_pred_linear, color = "red", linestyle = "--", label = "Predicted Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.title("Linear Regression : Marks Prediction")
plt.legend()
plt.grid(True)
plt.show()
```
